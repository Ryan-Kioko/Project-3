```python
#
import matplotlib.pyplot as plt

# Create a figure for the General ML Pipeline
plt.figure(figsize=(8, 14))  # Increased height to accommodate multi-line descriptions

# Define pipeline steps (9-step CRISP-DM structure) with multi-line descriptions
steps = [
    ("Dataset", "Audio: WAV files (Google Drive)\nReddit: Comments (PRAW API, 2024 posts)\nWeb: Articles (BeautifulSoup, predefined URLs)"),
    ("Data Preparation", "Audio: Transcribe (Whisper), Translate (googletrans), Split Sentences\nReddit: Scrape (PRAW), Detect Language, Translate (Helsinki-NLP)\nWeb: Scrape, Clean Text, Split Sentences"),
    ("Data Tagging", "Audio: Sentiment (DistilBERT), Labels: POSITIVE, NEGATIVE, NEUTRAL\nReddit: Sentiment (RoBERTa), Labels: POSITIVE, NEUTRAL, NEGATIVE\nWeb: Sentiment (DistilBERT), Labels: POSITIVE, NEGATIVE"),
    ("Model Building", "Audio/Web: DistilBERT (distilbert-base-uncased-finetuned-sst-2-english)\nReddit: RoBERTa (twitter-roberta-base-sentiment)\nPre-trained Hugging Face pipelines"),
    ("Model Prediction", "Audio: Sentiment per sentence chunk\nReddit: Sentiment per comment\nWeb: Sentiment per sentence\nOutput: Sentiment labels, scores"),
    ("Model Evaluation", "Metrics: Accuracy, Precision, Recall, F1"),
    ("Overall Sentiment Analysis", "Data Ingestion: Load CSVs\nPreprocessing: Transform scores (-1 to 1)\n\nVisualization: Bar plot, heatmap, word cloud\nOutput: finalproject.csv")
]

# Positions for each step (9 steps, vertically aligned, with more spacing for multi-line text)
y_positions = [0.94, 0.84, 0.74, 0.64, 0.54, 0.44, 0.34,]
x_position = 0.5

# Draw boxes and text for each step
for (title, description), y in zip(steps, y_positions):
    # Combine title and description into a single string with a newline
    text = f"{title}\n{description}"
    plt.text(x_position, y, text, ha='center', va='center', fontsize=9,
             bbox=dict(facecolor='#D1E8F0', edgecolor='#A3C1D3', boxstyle='round,pad=0.7', linewidth=1))

# Draw straight arrows between every pair of consecutive steps
# for i in range(len(y_positions) - 1):
#     plt.arrow(x_position, y_positions[i] - 0.07, 0, -0.05, head_width=0.02, head_length=0.02, fc='black', ec='black')

# Customize the plot
plt.title('General ML Pipeline for Sentiment Analysis', fontsize=16, pad=20)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.axis('off')  # Hide axes
plt.show()





```
