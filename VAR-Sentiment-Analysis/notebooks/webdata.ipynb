```python
import requests
from bs4 import BeautifulSoup

url= "https://talksport.com/football/2842313/jamie-ohara-var-investigation-arsenal-red-card/"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# scraping
paragraphs = soup.find_all('p')
text = ' '.join([para.get_text() for para in paragraphs])
print(text)

# cleaning
def clean_text(text):
    text = re.sub(r'[^A-Za-z0-9.,!?\s]', '', text)  # remove special characters
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra whitespace
    return text

import re
sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', text)
cleaned_sentences = [clean_text(sentence) for sentence in sentences]
cleaned_sentences
```

```python
# assign indices to sentences
indexed_sentences = {}
for index, sentence in enumerate(cleaned_sentences):
    indexed_sentences[index] = sentence

indexed_sentences
```

```python
# removing irrelevant sentences
indices_to_remove = [33, 45, 46, 47]

filtered_sentences = {}
for index, sentence in indexed_sentences.items():
    if index not in indices_to_remove:
        filtered_sentences[index] = sentence

# put the sentences back into a list
filtered_sentences_list = list(filtered_sentences.values())
filtered_sentences_list

# assign indices to the new list
indexed_sentences = {}
for index, sentence in enumerate(filtered_sentences_list):
    indexed_sentences[index] = sentence

indexed_sentences
```

```python
from transformers import pipeline

# declare model
model_name = "distilbert-base-uncased-finetuned-sst-2-english"

classifier = pipeline("sentiment-analysis", model=model_name)

# test on first 5 sentences
results = classifier(list(indexed_sentences.values())[:5])

for i, result in enumerate(results):
    print(f"Sentence {i}: {indexed_sentences[i]}")
    print(f"Sentiment: {result['label']}, Score: {result['score']}")
    print("-" * 20)
```
